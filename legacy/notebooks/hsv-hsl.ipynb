```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Image reading

```python
image1_path = "/home/adlane/projets/twinit-dataset/cards/90/a.jpg"
image2_path = "/home/adlane/projets/twinit-dataset/cards/90/b.jpg"
image1 = cv2.cvtColor(cv2.imread(image1_path, cv2.IMREAD_UNCHANGED), cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(cv2.imread(image2_path, cv2.IMREAD_UNCHANGED), cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
```

# RGB

```python
colors = ("red", "green", "blue")

# create the histogram plot, with three lines, one for
# each color
plt.figure()
fig, ax = plt.subplots(2)
for i, im in enumerate([image1, image2]):
    ax[i].set_xlim([0, 256])
    for channel_id, color in enumerate(colors):
        histogram, bin_edges = np.histogram(
            im[:, :, channel_id], bins=256, range=(0, 256)
        )
        ax[i].plot(bin_edges[0:-1], histogram, color=color)

    ax[i].set_title(f"Image {i+1}")
```

# HSV

```python
image1_hsv = cv2.cvtColor(image1, cv2.COLOR_RGB2HSV)
image2_hsv = cv2.cvtColor(image2, cv2.COLOR_RGB2HSV)
```

```python
colors = ("red", "green", "blue")

# create the histogram plot, with three lines, one for
# each color
plt.figure()
fig, ax = plt.subplots(2)
for i, im in enumerate([image1_hsv, image2_hsv]):
    ax[i].set_xlim([0, 256])
    for channel_id, color in enumerate(colors):
        max_value = 256 if channel_id > 0 else 180
        histogram, bin_edges = np.histogram(
            im[:, :, channel_id], bins=max_value, range=(0, max_value)
        )
        ax[i].plot(bin_edges[0:-1], histogram, color=color)

    ax[i].set_title(f"Image {i+1}")
```
