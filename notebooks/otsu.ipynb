```python
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(img):
    plt.figure(figsize=(20, 10))
    plt.imshow(img, cmap="gray")
    plt.show()
```

```python
image_path = Path(r"E:\OneDrive\Projets\twinit-dataset\28.jpg")
image = cv2.imread(str(image_path))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
show(gray)
```

```python
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
show(thresh)
```

```python
ret,th1 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,15,2)
th3 = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [gray, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imsave(f"{titles[i]}.jpg", images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
